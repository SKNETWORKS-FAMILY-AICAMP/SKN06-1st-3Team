```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
from selenium.webdriver.chrome.options import Options
options = ChromeOptions()
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

driver = webdriver.Chrome() # 실행위치와 동일한 곳에 chromedriver 위치하지 않으면 경로 지정 필요
from bs4 import BeautifulSoup
import requests

url = "https://www.hyundai.com/kr/ko/e/customer/center/faq"
driver.get(url)
# driver.maximize_window()
time.sleep(2)
question1 = []
answer1 = []
category1 = []
# driver.implicitly_wait(5)
wait = WebDriverWait(driver, 10)
for b in range(2,10):
    try : 
        try :
            for i in range(20):
                question_elements = driver.find_elements(By.CSS_SELECTOR, "span.list-content")
                qstn = [element.text.strip() for element in question_elements]
                for q in qstn:
                    question1.append(q)
                extend_buttons = driver.find_elements(By.XPATH, "//button[@class=\"list-title\"]")
                for ext in extend_buttons:
                    wait.until(EC.element_to_be_clickable(ext))
                    time.sleep(0.5)
                    ext.click()
                    # time.sleep(0.3)
                    category_elements = driver.find_elements(By.XPATH, '//button[@title="콘텐츠 확장됨"]/div/span[1]')
                    ctgr = [ca.text.strip() for ca in category_elements]
                    for c in ctgr:
                        category1.append(c)
                    answer_elements = driver.find_elements(By.XPATH, '//button[@title="콘텐츠 확장됨"]/following-sibling::div[1]/div')
                    ans = [an.text for an in answer_elements]
                    for a in ans:
                        answer1.append(a)
                    # time.sleep(0.3)
                next = driver.find_element(By.XPATH, "//li[contains(@class,'number active')]/following-sibling::li[1]/button")
                wait.until(EC.element_to_be_clickable(next))
                time.sleep(0.5)
                next.click()
                # time.sleep(0.3)
        except Exception as e:
            print("완료 또는 에러" , e)
            prev = driver.find_element(By.CSS_SELECTOR, "button.btn.btn-prevall")
            wait.until(EC.element_to_be_clickable(prev))
            time.sleep(0.5)
            prev.click()
            # time.sleep(0.3)
        ctgr_next = driver.find_element(By.XPATH, f'//*[@id="app"]/div[3]/section/div[2]/div/div[2]/section/div/div[1]/div[1]/ul/li[{b}]/button')
        wait.until(EC.element_to_be_clickable(ctgr_next))
        time.sleep(0.5)
        ctgr_next.click()
        # time.sleep(0.3)
    except ElementClickInterceptedException as e :
        print(type(e))
```
