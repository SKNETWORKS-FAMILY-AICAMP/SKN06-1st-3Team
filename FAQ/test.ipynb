```python
import requests
from bs4 import BeautifulSoup
import configparser
# 크롤링
# Config 파일 읽기
config = configparser.ConfigParser()
config.read('config.ini')  # config.ini 파일을 읽음

# Config에서 크롤링 설정값 가져오기
url = config['CRAWLING_Gi']['url']
user_agent = config['CRAWLING_Gi']['user_agent']
def crawl_data():
    # 요청 헤더 설정
    headers = {"user-agent": user_agent}
    
    # HTTP GET 요청 보내기
    res = requests.get(url, headers=headers)
    if res.status_code == 200:
        soup = BeautifulSoup(res.text, "lxml")
        faq_list = soup.select("div")
        # 추출한 데이터 넣어줄 배열
        
        search_names = []
        for txt in faq_list:
            # 태그 내의 텍스트 추출 후 공백 제거
            search_names.append(txt)
    print(search_names)
crawl_data()
```
